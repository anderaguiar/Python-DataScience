```python
#lista de salários
salarios = [900, 1200, 1500, 800, 12587, 10000, 600]
print(salarios)
```

```python
# verificar se existe o valor de salário igual a 10000
if 10000 in salarios:
    print('contido.')
else:
    print('não contido.')
```

```python
# imprimir o menor e o maior valor da lista de salários
print('menor valor: ', min(salarios))
print('maior valor: ', max(salarios))
```

```python
# adicione o valor 7300 à lista de salários
salarios.append(7300)
print(salarios)
```

```python
# extender a lista com dois novos valores utilizando apenas um método
valores = [5000, 6000]
salarios.extend(valores)
print(salarios)
```

```python
# imprimir o índice da lista de salários do elemento que possui o valor igual a 600
print('índice: ', salarios.index(600))
```

```python
# fazer uma ordenação dos valores da lista de salários, crescente e decrescente
sal_ord_crescente = sorted(salarios, reverse=False)
sal_ord_decrescente = sorted(salarios, reverse=True)
print('crescente:   ', sal_ord_crescente)
print('decrescente: ', sal_ord_decrescente)
```

```python
# baseado em uma imagem feita em aula ... 
arq_png = "angelina_jolie_cinza.png"

from IPython.display import Image
Image(filename = arq_png, width=300, height=300)
```

```python
# ... criar uma matriz do mesmo tamanho da imagem com valores igual a zero ...
import numpy as np
matriz = np.zeros((300,300))
print(matriz)
print("shape matriz: ", matriz.shape)
```

```python
# ... insira nesta lista os valores referente aos contidos na imagem
from PIL import Image
filename = arq_png
img = Image.open(filename)
img.load()
matriz = np.asarray(img, dtype="int32")
print("shape matriz: ", matriz.shape)
matriz_img = np.array(matriz).tolist()
arr_img = np.array(matriz_img)
print("shape arr_img: ", arr_img.shape)
arr_img
```

```python
# apresente a media, moda, maior valor, menor valor e o desvio padrão dessa matriz 
# transformada em um vetor

import statistics
```

```python
arr = arr_img.reshape(-1)
print("vetor: ", arr)
print("vetor_size: ", arr.size)
arr
```

```python
moda = statistics.mode(arr)
print("moda dos dados foi: ", moda)
```

```python
media = statistics.mean(arr)
print("média dos dados foi: ", media)
```

```python
print('menor valor foi: ', min(arr))
print('maior valor foi: ', max(arr))
```

```python
desvio = statistics.stdev(arr)
print("desvio padrao foi: ", desvio)
```
